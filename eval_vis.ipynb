```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from adjustText import adjust_text

mal = pd.read_csv('./evals/eval_mean_angular_loss.csv')
al = pd.read_csv('./evals/eval_angular_loss.csv')
```

```python
mean_value = mal.mean(axis=1).values

# Reshape the dataframe for seaborn
df_melted = mal.melt(var_name="Batch", value_name="Value")

# Plot the data
fig, ax = plt.subplots(figsize=(10,5))

# Plot the point plot using seaborn
sns.pointplot(x="Batch", y="Value", data=df_melted, ax=ax, join=False)

# Plot the mean line
ax.axhline(mean_value, color="red", linestyle="--", label="Mean")

# Customize the plot
ax.set_xlabel("Batches")
ax.set_ylabel("Mean Angular Loss")
ax.legend()

# Add non-overlapping labels
texts = [ax.text(p[0], p[1], p[1], fontsize=9) for p in zip(range(len(mal.columns)), mal.iloc[0])]
adjust_text(texts, arrowprops=dict(arrowstyle="-", color='grey', lw=0.5))

# Rotate x-axis labels
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
